Load Libraries

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
def gradientDescent(X, theta, Y, learning_rate):
    m, n = X.shape
    slope = (np.dot(np.transpose(X),np.dot(X,theta)-Y)) / (m)
    theta = theta - learning_rate * slope
    return theta
```

```python
def costFunction(X, theta, Y):
    m, n = X.shape
    cost = np.dot(X, theta) - Y;
    cost = np.dot(np.transpose(cost), cost) / X.shape[0]
    cost = np.sum(cost)
    return cost
```

```python
#get dataset
dataset = pd.read_csv("house_price.csv");
print(dataset.head())
data = np.array(dataset)
X = data[:4600,1:]   # m x n 
Y = data[:4600,:1]   # m x 1
theta = np.zeros((X.shape[1], 1)) # n x 1
print(X.shape)
print(Y.shape)
print(theta.shape)
for i in range(X.shape[0]):
    X[:][i] = X[:][i] / X[:][i].max()
#Y = Y / Y.max()
```

Iterations

```python
epochs = 5000
learning_rate = 0.01
cost = []
for i in range(epochs):
    theta = gradientDescent(X, theta, Y, learning_rate)
    cost.append(costFunction(X, theta, Y))
predicted = np.dot(X, theta)
    
```

```python
plt.plot(np.arange(0, epochs), cost)
plt.title("#Iterations Vs Cost")
plt.xlabel("#Iterations")
plt.ylabel("Cost")
plt.show()
print(cost[:5])
```

```python
Xaxis = np.arange(0, 100)
plt.plot(Xaxis, Y[:100,:], "b.", label = "Expected")
plt.plot(Xaxis, predicted[:100,:], "r.", label = "Predicted")
plt.title("Expected VS Predicted")
plt.legend()
plt.show()
print(Y)
```
